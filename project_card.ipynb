```python
import pandas as pd
df = pd.read_excel('test.xlsx')

df['Column Money Out (£).'] = (df['Column Money Out (£).'].str.replace("Money Out (£)", "", regex=False)
    .apply(lambda x: x[:-1] if x.endswith('.') else x)
    .str.strip()
    .str.replace("blank","0", regex=False)
    .str.replace(",","", regex=False)
    .astype(float))

df['Column Money In (£).'] = (df['Column Money In (£).'].str.replace("Money In (£)", "", regex=False)
    .apply(lambda x: x[:-1] if x.endswith('.') else x)
    .str.replace("blank","0", regex=False)
    .str.strip()
    .str.replace(",","", regex=False)
    .astype(float))

df['Column Balance (£).'] = (df['Column Balance (£).'].str.replace("Balance (£)", "", regex=False)
    .apply(lambda x: x[:-1] if x.endswith('.') else x)
    .str.strip()
    .str.replace(",","", regex=False)
    .astype(float))

df['Column Type.'] = (df['Column Type.'].str.replace("Type", "", regex=False)
    .str.replace(".", "", regex=False)
    .str.strip())

df['Column Description.'] = (df['Column Description.'].str.replace("Description", "", regex=False)
    .str.replace(".", "", regex=False)
    .str.strip())

df['Column Date.'] = pd.to_datetime(df['Column Date.'].str.replace("Date", "", regex=False)
    .str.replace(".", "", regex=False)
    .str.strip(), format="%d %b %y", errors="coerce")
```

```python
df.to_csv("PersonalFinanceHist.csv", index= False)
```
